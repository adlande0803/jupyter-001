HERE THE OUT PUT OF OPRATION PERFOERM ON THE NUMPY ARRAY  AND THIS PROCESES IS CALLED AS VECTORISATION

```python
LETS SAY IF WE HAVE TO DO OPARTION ON NUMPY ARRAY AND IN THAT OPRATION
WE RETURN THE SQUARE OF ELEMENT IF IT IS EVEN AND CUBE IF IT IS ODD. 
```

```python
a=np.arange(10)
a
```

```python
def ops(z):
    if z%2==0:
        return z*z
    elif z%2==1
        return z*z*z
```

```python
ops(8)
```

```python
ops(7)
```

# BUT WE CAN'T USE THIS FUNCTION ON NUMPY ARRAY SO TO DO THAT

```python

HERE WE GOING TO BUILD NEW FUNCTION WHICH IS CAPABLE TO TAKE NUMPY ARRAY AS THE ARGUMENT 
```

```python
# IN NEXT STEP WE USE VECTORIZE FUCTION FROM NUMPY TO MAKE MAGIC FUNCTION WHICH WILL CAN ACCEPT THE ARRAY
```

```python
MAGIC = np.vectorize(ops)
```
